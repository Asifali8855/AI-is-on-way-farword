```python
from huggingface_hub import login
from esm.models.esm3 import ESM3
from esm.sdk.api import ESMProtein, GenerationConfig

# login to Hugging Face to accept license / get access
login()
# load the open-model (small version) locally
model = ESM3.from_pretrained("esm3_sm_open_v1").to("cuda")  # or "cpu"

prompt = "____AAACCCGGG____"
protein = ESMProtein(sequence=prompt)
protein = model.generate(protein, GenerationConfig(track="sequence", num_steps=8, temperature=0.7))
protein = model.generate(protein, GenerationConfig(track="structure", num_steps=8))
protein.to_pdb("out.pdb")
```

```python
from esm.models.esm3 import ESM3
from esm.sdk.api import ESMProtein, GenerationConfig

# Load model (open small version for example) and cast to float
model = ESM3.from_pretrained("esm3_sm_open_v1").to("cpu").float()

# Wrap your generated sequence
protein = ESMProtein(sequence="AAAAAAACCCGGGTTTTTCCCCCCAAAAUUUUUUUCLLL")

# Ask model to predict structure
protein = model.generate(protein, GenerationConfig(track="structure", num_steps=8))

# Save as PDB file
protein.to_pdb("generated_protein.pdb")
```
